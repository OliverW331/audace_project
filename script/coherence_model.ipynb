```python
from gensim.test.utils import common_corpus, common_dictionary
from gensim.models.ldamodel import LdaModel
from gensim.models.coherencemodel import CoherenceModel
```

```python
import matplotlib.pyplot as plt
```

```python
print('Number of unique tokens: %d' % len(common_dictionary))
print('Number of documents: %d' % len(common_corpus))
```

```python
coherence_scores = []
perplexity_scores = []
for i in range(1, 100):
  model = LdaModel(
    corpus=common_corpus, 
    num_topics=i, 
    id2word=common_dictionary,
    iterations = 500,
    passes = 5,
    alpha = 'auto',
    eta = 'auto',)
  top_topics = model.top_topics(common_corpus)
  coherence_model_lda = CoherenceModel(model=model, corpus=common_corpus, coherence='u_mass')
  coherence_lda = coherence_model_lda.get_coherence()
  print('nCoherence Score: ', coherence_lda)
  coherence_scores.append(coherence_lda)
```

```python
plt.plot(range(1,100), coherence_scores)
plt.title("The average coherence scores of different numbers of topics")
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score")
plt.show
```

```python
from gensim.topic_coherence import direct_confirmation_measure, text_analysis
from collections import namedtuple
```

```python
# Create dictionary
id2token = {1: 'test', 2: 'doc'}
token2id = {v: k for k, v in id2token.items()}
dictionary = namedtuple('Dictionary', 'token2id, id2token')(token2id, id2token)
```

```python
dictionary
```

```python
# Initialize segmented topics and accumulator
segmentation = [[(1, 2)]]

accumulator = text_analysis.InvertedIndexAccumulator({1, 2}, dictionary)
```

```python
accumulator.relevant_words
```

```python
accumulator._inverted_index = {0: {2, 3, 4}, 1: {3, 5}}
accumulator._num_docs = 5
```

```python
# result should be ~ ln{(1 / 5) / [(3 / 5) * (2 / 5)]} = -0.182321557
result = direct_confirmation_measure.log_ratio_measure(segmentation, accumulator)[0]
```
